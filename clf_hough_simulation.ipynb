```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
np.set_printoptions(threshold=np.inf)  # set the numpy print options so that all the array values are shown
np.set_printoptions(linewidth=np.inf)  # set the numpy print options so that all the array values are shown
```

```python
def noise_removal(pixel_sector, kernal_size, threshold):
    pixel_initial = pixel_sector.copy().reshape(32,32)
    pixel_cleaned = pixel_initial.copy()
    length = int((kernal_size-1)/2)
    nonzero_indices = pixel_initial.nonzero()
    for i in range(nonzero_indices[0].size):
        index_x, index_y = nonzero_indices[0][i], nonzero_indices[1][i]
        x1, x2 = max(0, index_x - length), min(31, index_x + length)
        y1, y2 = max(0, index_y - length), min(31, index_y + length)
        if (pixel_initial[x1:(x2+1), y1:(y2+1)].sum()<threshold):
            pixel_cleaned[index_x, index_y] = 0
    return pixel_cleaned.reshape(1024)
```

```python
def hough_clf(pixel_sector, para1, para2):
    hough_input = np.array(pixel_sector.reshape(32,32)*255, dtype=np.uint8)
    hough_output = cv.HoughCircles(hough_input,cv.HOUGH_GRADIENT, dp=1, minDist=100,\
                                   param1=para1, param2=para2, minRadius=5, maxRadius=12)
    hough_return = np.empty(3)
    if (np.size(hough_output)==1):
        hough_return[0] = 0
        hough_return[1] = 0
        hough_return[2] = 0
    else:
        hough_return[0] = hough_output[0][0][2]
        hough_return[1] = hough_output[0][0][0]
        hough_return[2] = hough_output[0][0][1]
    
    return hough_return
```

```python
def plot_circle(hough_return):
    
    plot = np.zeros((32,32))
    if (hough_return[0]>0):
        for i in range(200):
            circx = hough_return[1] + np.cos(i/200*2*np.pi)*hough_return[0]
            circy = hough_return[2] + np.sin(i/200*2*np.pi)*hough_return[0]
            if (circx>31) or (circy>31) or (circx<0) or (circy<0):
                continue
            plot[int(np.around(circy)),int(np.around(circx))] = 5
        
    return plot.reshape(1024)
```
